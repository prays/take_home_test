# Preprocessing Data

```python
df_one_hot = pd.get_dummies(df, columns=['STATUS_OF_EXISTING_CHECKING_ACCOUNT','CREDIT_HISTORY','PURPOSE','SAVINGS_ACCOUNT/BONDS','PRESENT_EMPLOYMENT_SINCE','PERSONAL_STATUS_AND_SEX','OTHER_DEBTORS/GUARANTORS',
      'PROPERTY','OTHER_INSTALLMENT_PLANS','HOUSING','JOB','TELEPHONE','FOREIGN_WORKER'])
```

```python
df_one_hot.head()
```

df_one_hot adalah dataframe baru dimana data kategorik sudah diinialisasi sebagai 0 dan 1 untuk tiap kategorik yang muncul atau tidak, sehingga kolom menjadi jumlah kolom menjadi besar dikarenakan banyaknya kategorik.

# Prediction Algorithm

```python
from sklearn.model_selection import train_test_split
target = ['CREDIT_SCORE']
X = df_one_hot.drop(target,axis=1)
Y = df_one_hot['CREDIT_SCORE']
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state=95)
```

Membagi data menjadi train dan test 

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

Akan digunakan Regresi Logistik sebagai Machine Learning

```python
prediction = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```

Yang berhasil diprediksi sebanyak 157 dan yang salah diprediksi sebanyak 42 sehingga akurasinya 75%
